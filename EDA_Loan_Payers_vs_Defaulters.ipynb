#By just analysing single variables, we won't find much insights related to the defaulters, as here we will just have an idea which category of people are present in abundance, 
#other than that, most of the insights are gathered in analysis of multiple features/variables with target variable.

```python
def bivariate(feat):
    d1 = cur_app[[feat, 'TARGET']].groupby([feat],as_index=False).mean()
    d1.sort_values(by='TARGET', ascending=False, inplace=True)

    fig,((ax1,ax2))=plt.subplots(1,2, figsize=(15,6))
    sns.barplot(x=cur_app[feat].value_counts().index, y=cur_app[feat].value_counts().values, ax=ax1)
    ax1.set_title(feat)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
    sns.barplot(x=feat, y='TARGET', order=d1[feat],data=d1, ax=ax2)
    ax2.set_title(feat)
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=90)
    plt.ylabel('% of applications where loans were NOT repayed', fontsize=10)   
    plt.tight_layout(pad=2)
    plt.show()
```

```python
for i in cols:
    bivariate(i)
```

```python
def univariate_prev(feat):
    d1 = prev_app[feat].value_counts()*100/prev_app.shape[0]
    d1.sort_values (ascending=False, inplace=True)

    fig,((ax1,ax2))=plt.subplots(1,2, figsize=(10,6))
    sns.barplot(x=prev_app[feat].value_counts().index, y=prev_app[feat].value_counts().values, ax=ax1)
    ax1.set_title(feat)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
    sns.barplot(x=d1.index, y=d1.values, ax=ax2)
    ax2.set_title(feat)
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=90)
    plt.ylabel('% of applications where loans were NOT repayed', fontsize=10)   
    plt.tight_layout(pad=2)
    plt.show()
```
