```python
#now both datasets have country code in common so we will combine both of them and get in one df using pd.merge on country code variable
final_df = pd.merge(df, df_country, on= 'Country Code', how= 'left')
final_df.head() #pd.concat can also be used
```

```python
#Lets get some insights 
#now we see we have a col, called country which =basically tells which are the countries in which zomato is used
index = final_df.Country.value_counts().index #its tells INdia is the base of zomato more transactions happen here...bcoz US just has  a website which suggest places
#but most of transcations/ orders happen in India(base of Zomato)
```

```python
values = final_df.Country.value_counts().values # tjhis gives values of num. of records for countries
```

```python
#next we see some cols for ratings so lets see from which country we get max ratings ...groupby rating cols and see records for them ...which is max rating
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0 : 'Rating_count'})
#.size gicves num of records per groubpy category 
ratings
```

```python
ratings['Rating_count'].sum()
```

## some obs :

3. when ratiungs b/w 4 -4.4 --> Very Good service
4. ratings b/w 3.5 and 3.9 --> Good sevice

```python
#lets plot this obs
#which rating is given max times
plt.rcParams['figure.figsize'] =(12,6)
sns.barplot(x='Aggregate rating', y='Rating_count', data =ratings)
plt.show()
```
